```python
import pandas as pd
import numpy as np
```

```python
movie = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movie.head()
```

```python
credits.head(1)
```

```python
movies = movie.merge(credits, on='title')
```

```python
movies.info()
```

```python
#genres keywords id  overview title cast  crew

movies = movies[['genres','movie_id','keywords','title','overview','cast','crew']]
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace = True)
```

```python
movies.isnull().sum()
```

```python
movies.iloc[0]['genres']
```

```python
import ast

def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):   #to convert obj into list
        l.append(i['name'])
    return l
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head(1)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head(1)
```

```python
def conv(obj):
    c=0
    l=[]
    for i in ast.literal_eval(obj):
        if c!=3:
            l.append(i['name'])
            c+=1
        else:break
    return l


```

```python
movies['cast']
```

```python
movies.head(1)
```

```python
def director(obj):
    l = []
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            l.append(i['name'])
            break
    return l
```

```python
movies['crew']=movies['crew'].apply(director)
```

```python
movies.head(1)
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(' ','') for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(' ','') for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(' ','') for i in x])
```

```python
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(' ','') for i in x])
```

```python
movies['tags'] = movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
new_dataset = movies[['movie_id','title','tags']]
new_dataset.head()
```

```python
new_dataset['tags']= new_dataset['tags'].apply(lambda x:' '.join(x))
```

```python
new_dataset.head()
```

```python
new_dataset['tags']= new_dataset['tags'].apply(lambda x:x.lower())
```

```python
new_dataset.head()
```

```python
#stemming

import nltk
from nltk.stem.porter import PorterStemmer

p = PorterStemmer()
```

```python

def stem(text):
    l = []
    for i in text.split():
        l.append(p.stem(i))
    
    return ' '.join(l)    
```

```python
new_dataset['tags'][0]

text = 'in the 22nd century, a paraplegic marine is dispatched to the moon pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. action adventure fantasy sciencefiction cultureclash future spacewar spacecolony society spacetravel futuristic romance space alien tribe alienplanet cgi marine soldier battle loveaffair antiwar powerrelations mindandsoul 3d jakesully neytiri dr.graceaugustine jamescameron'
print(stem(text))

new_dataset['tags'] = new_dataset['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_dataset['tags']).toarray()
```

```python
cv.get_feature_names()


#we need to apply stemming
#['love','loving'] =>stemming = ['love',love]
```

```python
#Similarity between movies

from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)

print(similarity)

```

```python

def recommend(movie):
    movie_ind = new_dataset[new_dataset['title']==movie].index[0]
    
    
    movie_list = sorted(enumerate(similarity[movie_ind]),reverse=True , key = lambda x:x[1])[1:6]
    
    for i in movie_list:
        print(new_dataset.iloc[i[0]].title)
    
```
